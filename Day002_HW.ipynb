### 比較一下範例檔案中的「File I/O」與「CSV Reader」讀出來的內容有什麼差異

```python
#File I/O
fh = open("./data/example.csv", encoding="utf-8")
f = fh.read()
fh.close()
print(f)

#CSV Reader
import csv
# 開啟 CSV 檔案
with open('./data/example.csv', newline='', encoding="utf-8") as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.reader(csvfile)
    # 以迴圈輸出每一列
    for row in rows:
        print(row)
```

### 根據範例檔案的結果：

1. 取出班次一的每一個時間
2. 將班次一的每一個時間用一種資料型態保存
3. 將班次一到五與其所有時間用一種資料型態個別保存

```python
# 1. 取出班次一的每一個時間
list_1st=[]
with open('./data/example.csv', newline='', encoding="utf-8") as csvfile:
    rows = csv.reader(csvfile)
    
# 2. 將班次一的每一個時間用一種資料型態保存
    for row in rows:
        time = row[5]
        list_1st.append(time)
print(list_1st)
```

```python
# 3. 將班次二到五與其所有時間用一種資料型態個別保存
list_2nd=[]
with open('./data/example.csv', newline='', encoding="utf-8") as csvfile:
    rows = csv.reader(csvfile)
    for row in rows:
        time = row[6]
        list_2nd.append(time)
print(list_2nd)

list_3rd=[]
with open('./data/example.csv', newline='', encoding="utf-8") as csvfile:
    rows = csv.reader(csvfile)
    for row in rows:
        time = row[7]
        list_3rd.append(time)
print(list_3rd)

list_4th=[]
with open('./data/example.csv', newline='', encoding="utf-8") as csvfile:
    rows = csv.reader(csvfile)
    for row in rows:
        time = row[8]
        list_4th.append(time)
print(list_4th)

list_5th=[]
with open('./data/example.csv', newline='', encoding="utf-8") as csvfile:
    rows = csv.reader(csvfile)
    for row in rows:
        time = row[9]
        list_5th.append(time)
print(list_5th)
```
